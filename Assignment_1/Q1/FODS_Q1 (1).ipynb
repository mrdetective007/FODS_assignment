```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta 
```

# Explanation-1

The initial task is to plot the beta distribution using the initially provided parameters i.e., a=2 and b=2. 
This is the initial prior distribution with the provided parameters.

```python
a, b = 2, 2
x = np.linspace(0, 1, 1000)

plt.figure(figsize=(5,5)) 
plt.xlim(0, 1) 
plt.plot(x, beta.pdf(x, a, b), 'green') 
plt.title('Beta Distribution for the given problem')
plt.xlabel('Values of Random Variable X in the range of 0 and 1')
plt.ylabel('Probability Density for the given problem')
plt.show()
```
