```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
x1 = df["MLOGP"]
x2 = df["RDCHI"]
y = df["quantitative response of LC50"]
```

```python
def ferr(w1, w2):
    error = 0
    error = ((w1*x1 + w2*x2 - y)**2).sum(axis=0)
    error = error/2
    return error
```

```python
w1 = np.arange(-1,1,0.005)
w2 = np.arange(-1,1,0.005)
MSE_final = np.zeros((len(w1), len(w2)), float)
```

```python
for i in range (len(w1)):
    for j in range (len(w1)):
        MSE_final[i,j] = ferr(w1[i],w2[j]) 
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(20)
plt.contour(w1, w2, MSE_final,88,cmap='RdGy');
plt.colorbar();
plt.xticks(np.linspace(-1,1,30))
plt.yticks(np.linspace(-1,1,30))

x1 = np.linspace(-1, 1, 1000)
y1 = np.linspace(-1, 1, 1000)
a1, b1 = np.meshgrid( x1 , y1 )
C1 = 1.4 - (abs(a1)**0.5 + abs(b1)**0.5)
plt.contour( a1 , b1 , C1 , [0], colors = 'green')
plt.xlabel('w1')
plt.ylabel('w2')
plt.show()
```
